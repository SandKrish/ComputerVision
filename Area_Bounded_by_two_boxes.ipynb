This program calculates total area of the bounding boxes and finds the intersection of points of two boxes.

image.png

Taking 2 boxes boxA an boxB

```python
def box_intersection(boxA, boxB):
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[2], boxB[2])
    yB = min(boxA[3], boxB[3])

    if (xA > xB or yA > yB):
      print('No intersection')
    else:
      interArea = (xB - xA + 1) * (yB - yA + 1)
      print('Area of intersection is',interArea)    
      print('point of intersection is ', '(', xA, ',', yA ,')','(', xB , ',', yA ,')', '(',xB, ',', yB ,')', '(',xA, ',', yB,')')
```

```python
boxA = [39, 63, 203, 112]
boxB = [54, 66, 198, 114]  
```

```python
box_intersection(boxA,boxB)
```
